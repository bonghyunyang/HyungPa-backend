```python
bulk_list=[]
with open(file_loca+ 'product.csv') as hand:
    reader = csv.DictReader(hand)
    for row in reader:
        bulk_list.append(Product(
            brand_name = row['brand'],
            name = row['product'],
            product_image = row['product_image'],
            like_nums = row['like_number'],
            review_nums = row['review'],
            mini_nums = row['mini'],
            capacity = row['capacity'],
            price = row['price'],
            firstcategory_id = row['first_category'],
            secondcategory_id = row['second_category']))
Product.objects.bulk_create(bulk_list)
```
