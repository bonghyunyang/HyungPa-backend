```python
bulk_list = []
with open(file_loca + 'first_categories.csv') as hand:
    reader = csv.reader(hand) #csv 파일을 읽고 변수에 저장합니다.
    next(reader, None)
    for row in reader:
        bulk_list.append(FirstCategory(
            name=row[0]))
```

```python
FirstCategory.objects.bulk_create(bulk_list)
```

```python
bulk_list = []
with open(file_loca + 'product.csv') as hand:
    reader = csv.DictReader(hand) #csv 파일을 읽고 변수에 저장합니다.
#   next(reader, None)
    for row in reader:
        print(row['brand'])
        bulk_list.append(Product(
            brand_name = row['brand'], # brand
            name = row['product'],
            product_image = row['product_image'],
            like_nums = row['like_number'],
            review_nums = row['review'],
            mini_nums = row['mini'],
            capacity = row['capacity'],
            price = row['price'],
            firstcategory_id = row['first_category'],
            secondcategory_id = row['second_category']))
```

```python
bulk_list = []
with open(file_loca + 'post_crawling.csv') as hand:
    reader = csv.DictReader(hand) #csv 파일을 읽고 변수에 저장합니다.
    next(reader, None)
    for row in reader:
        print(row['title'])
        bulk_list.append(Post(
            title = row['title'], 
            description = row['description'],
            like_number = row['like_number'],
            view_number = row['view_count'],
            ))
```
